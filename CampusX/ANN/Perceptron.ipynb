```python
df=pd.read_csv('students.csv')

print(df.shape)
df.head()
```

```python
sn.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['isPlaced'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron #conda install scikit-learn
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_
# w1=-344.5,w2=56
```

```python
p.intercept_
# b=-108
```

```python
from mlxtend.plotting import plot_decision_regions #conda install -c conda-forge mlxtend
```

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```
