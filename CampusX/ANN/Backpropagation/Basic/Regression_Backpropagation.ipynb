```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5 ,12,7]],columns=['cgpa','profile_score','lpa'])
```

```python
#this function will get our model architecture i.e. [number_of_nodes_at_input_layer,number_of_nodes_hidden_layer,number_of_nodes_output_layer] 
# it creates total number of weights and biases for us with default values
def initialize_parameters(layers_dim):
    np.random.seed(3)
    parameters={}
    L=len(layers_dim)

    for l in range(1,L):
        parameters['W'+str(l)]=np.ones((layers_dim[l-1],layers_dim[l]))*0.1 # Default value of weights is 0.1
        parameters['b'+str(l)]=np.zeros((layers_dim[l],1)) # Default value of biases is 0

    return parameters
    
```

```python
initialize_parameters([2,2,1])

#W1 is the weights from input layer to first hidden layer since 4 connections are there so 4 values are there
#W2 is the weights from hidden layer to output layer since 2 weights are there so 2 values 
#b1 is the biases of hidden layer nodes i.e. 2 nodes
#b2 is the bias of the output layer node i.e. 1 node

```

```python
# this method will calculate the output for a given node

def linear_forward(A_prev,W,b):

    Z=np.dot(W.T,A_prev)+b #Linear activation Function

    return Z
```

```python
# Main Function which will perform forward Propagation
def L_layer_forward(X,parameters):
    A=X
    L=len(parameters)//2 #number of layers in the neural network

    for l in range(1,L+1):
        A_prev=A
        W1=parameters['W'+str(l)]
        b1=parameters['b'+str(l)]

        # print("A"+str(l-1)+": ",A_prev)
        # print("W"+str(l)+": ",W1)
        # print("b"+str(1)+": ",b1)

        # print("--" * 20)

        A=linear_forward(A_prev,W1,b1)
        # print("A"+str(l)+": ",A)
    

    return A,A_prev # it returns value of the final prediction and the values of the previous layer
```

```python
X=df[['cgpa','profile_score']].values[0].reshape(2,1) #we get array so values[0] 
y=df[['lpa']].values[0][0] # we get simple value number types so values[0][0]

parameters=initialize_parameters([2,2,1])
y_hat,A1=L_layer_forward(X,parameters)

# output will be 
# 1st
    # A0=Input data from the table 
    # W1=parameter we have prepared values are 0.1 and 4 values are there
    # b1= array of biases of both the nodes at hidden layer 1
#2nd
    #A1=data received after forward propagation from nodes in first hidden layer i.e the combination of outputs from both the nodes in Hidden-layer:1 is called A1
    #W2=weights b/w hidden-layer1 and output layer
    #b1= bias on the output layer node
#3rd
    #A1=Final Output
```

```python
def loss_function(y_hat):
    return (y-y_hat)**2
```

```python
#Update the Weights 
# Wnew=Wold-lr*derivative of loss_function wrt Wold
def update_parameters(parameters,y,y_hat,hidden_layer_output,X):
    learning_rate=0.001
    #Final output layer updation 
    parameters['W2'][0][0]=parameters['W2'][0][0]-(learning_rate*(-2*(y-y_hat))*hidden_layer_output[0][0])
    parameters['W2'][1][0]=parameters['W2'][1][0]-(learning_rate*(-2*(y-y_hat))*hidden_layer_output[1][0])
    parameters['b2'][0][0]=parameters['b2'][0][0]-(learning_rate*(-2*(y-y_hat)))

    #First Node on the Hidden Layer Updation
    parameters['W1'][0][0]=parameters['W1'][0][0]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][0][0]*X[0][0])
    parameters['W1'][0][1]=parameters['W1'][0][1]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][0][0]*X[1][0])
    parameters['b1'][0][0]=parameters['b1'][0][0]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][0][0])
    
    #Second Node on the Hidden Layer Updation
    parameters['W1'][1][0]=parameters['W1'][1][0]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][1][0]*X[0][0])
    parameters['W1'][1][1]=parameters['W1'][1][1]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][1][0]*X[1][0])
    parameters['b1'][1][0]=parameters['b1'][1][0]-(learning_rate*(-2*(y-y_hat))*parameters['W2'][1][0])
```

```python

update_parameters(parameters,y,y_hat[0][0],A1,X)
```

```python
#Epochs Implementation
parameters=initialize_parameters([2,2,1])
epochs=5

for i in range(epochs):
    Loss=[]
    for j in range(df.shape[0]):
        X=df[['cgpa','profile_score']].values[j].reshape(2,1)
        y=df[['lpa']].values[j][0]

        #parameter Initialization
        y_hat,A1=L_layer_forward(X,parameters)
        y_hat=y_hat[0][0]


        update_parameters(parameters,y,y_hat,A1,X)

        Loss.append((y-y_hat)**2)
    
    print("Epoch -",i+1,'Loss -',np.array(Loss).mean())

parameters
```
