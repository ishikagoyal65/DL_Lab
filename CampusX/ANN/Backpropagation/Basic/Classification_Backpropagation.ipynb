# Making Model for Classification type Problems
- 3 columns:(cgpa,profile_score,placement)
- Architecture is same i.e input layer(2 input nodes),hidden_layer(2 nodes),output_layer(1 node)
- In regression Example Activation function in all the layer was 'linear' here we will use 'Sigmoid' in all the layer as activation function
- Loss function is Binary Cross Entropy: -ylog(y_hat)-(1-y)log(1-y_hat)



```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]],columns=['cgpa','profile_score','placed'])
```

```python
def initialize_parameters(layer_dims):
    np.random.seed(3)
    parameters={}
    L=len(layer_dims)

    for l in range(1,L):
        parameters['W'+str(l)]=np.ones((layer_dims[l-1],layer_dims[l]))*0.1 #Default value of 0.1 to all the weights
        parameters['b'+str(l)]=np.zeros((layer_dims[l],1))# Default value of 0 to all the biases

    
    return parameters
```

```python
#Sigmoid Activation Function

def sigmoid(z):
    A=1/(1+np.exp(-z))

    return A
```

```python
def linear_forward(A_prev,W,b):
    z=np.dot(W.T,A_prev)+b
    after_activation=sigmoid(z) #Sigmoid Activation Function

    return after_activation
```

```python
# Main Function which will perform forward Propagation
def L_layer_forward(X,parameters):
    A=X
    L=len(parameters)//2 #number of layers in the neural network

    for l in range(1,L+1):
        A_prev=A
        W1=parameters['W'+str(l)]
        b1=parameters['b'+str(l)]

        # print("A"+str(l-1)+": ",A_prev)
        # print("W"+str(l)+": ",W1)
        # print("b"+str(1)+": ",b1)

        # print("--" * 20)

        A=linear_forward(A_prev,W1,b1)
        # print("A"+str(l)+": ",A)
    

    return A,A_prev # it returns value of the final prediction and the values of the previous layer
```

```python
def update_parameters(parameters,y,y_hat,hidden_layer_output,X):
    learning_rate=0.0001
    parameters['W2'][0][0]=parameters['W2'][0][0]-(-learning_rate*(y-y_hat)*hidden_layer_output[0][0])
    parameters['W2'][1][0]=parameters['W2'][1][0]-(-learning_rate*(y-y_hat)*hidden_layer_output[1][0])
    parameters['b2'][0][0]=parameters['b2'][0][0]-(-learning_rate*(y-y_hat))

    parameters['W1'][0][0]=parameters['W1'][0][0]-(-learning_rate*(y-y_hat)*parameters['W2'][0][0]*hidden_layer_output[0][0]*(1-hidden_layer_output[0][0])*X[0][0])
    parameters['W1'][0][1]=parameters['W1'][0][1]-(-learning_rate*(y-y_hat)*parameters['W2'][0][0]*hidden_layer_output[0][0]*(1-hidden_layer_output[0][0])*X[1][0])
    parameters['b1'][0][0]=parameters['b1'][0][0]-(-learning_rate*(y-y_hat)*parameters['W2'][0][0]*hidden_layer_output[0][0]*(1-hidden_layer_output[0][0]))

    parameters['W1'][1][0]=parameters['W1'][1][0]-(-learning_rate*(y-y_hat)*parameters['W2'][1][0]*hidden_layer_output[1][0]*(1-hidden_layer_output[1][0])*X[0][0])
    parameters['W1'][1][1]=parameters['W1'][1][1]-(-learning_rate*(y-y_hat)*parameters['W2'][1][0]*hidden_layer_output[1][0]*(1-hidden_layer_output[1][0])*X[1][0])
    parameters['b1'][1][0]=parameters['b1'][1][0]-(-learning_rate*(y-y_hat)*parameters['W2'][1][0]*hidden_layer_output[1][0]*(1-hidden_layer_output[1][0]))

    
```

```python
# first epoch propagation
X=df[['cgpa','profile_score']].values[0].reshape(2,1)
y=df[['placed']].values[0][0]

parameters=initialize_parameters([2,2,1])

y_hat,hidden_layer_output=L_layer_forward(X,parameters)

y_hat=y_hat[0][0]

update_parameters(parameters,y,y_hat,hidden_layer_output,X)

print("Loss of the Student-",(-y*np.log(y_hat)-(1-y)*np.log(1-y_hat)))

parameters

```

```python
#Epochs Implementation
parameters=initialize_parameters([2,2,1])
epochs=50

for i in range(epochs):
    Loss=[]
    for j in range(df.shape[0]):
        X=df[['cgpa','profile_score']].values[j].reshape(2,1)
        y=df[['placed']].values[j][0]

        #parameter Initialization
        y_hat,A1=L_layer_forward(X,parameters)
        y_hat=y_hat[0][0]


        update_parameters(parameters,y,y_hat,A1,X)

        Loss.append((-y*np.log(y_hat)-(1-y)*np.log(1-y_hat)))
    
    print("Epoch -",i+1,'Loss -',np.array(Loss).mean())

parameters
```
