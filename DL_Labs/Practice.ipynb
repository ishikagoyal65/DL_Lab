```python
import numpy as np
import matplotlib.pyplot as pt
```

```python
c=10
Da=20
Dx=200
m=100

X=np.random.randn(m,Dx)
Y=np.zeros((m,c))
y_labels=np.random.randint(0,c,size=m)

Y[np.arange(m),y_labels]=1 

print(f"XShape:{X.shape}, YShape:{Y.shape}")
```

```python
def softmax(z):
    exps=np.exp(z)
    return exps/np.sum(exps)
```

```python

W1=np.random.randn(Da,Dx)
b1=np.zeros((Da,1))
W2=np.random.randn(c,Da)
b2=np.zeros((c,1))

X_transpose=X.T # Transposing to match the shape so that it can be multiplied

z1=np.dot(W1,X_transpose)+b1
a1=np.tanh(z1)

z2=np.dot(W2,a1)+b2
y_hat=softmax(z2)

 
```

```python
Y_transpose=Y.T
loss_Function=-np.sum(Y_transpose*np.log(y_hat+1e-9))/m

print("Loss Function:",loss_Function)
```
