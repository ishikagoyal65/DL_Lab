```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dx=200 #input rows
c=10 #final classes 
Da=20 # rows in hidden layer
m=100 #columns of data

X=np.random.randn(Dx,m)
```

```python
#One-hot Encoding
y_labels=np.random.randint(0,c,size=m) #100 random integers as our dummy output
Y=np.zeros((m,c))
Y[np.arange(m),y_labels]=1
```

```python


W1 = np.random.randn(Da, Dx) *0.01
b1 = np.zeros((Da,1))*0.01

W2 = np.random.randn(c, Da) *0.01
b2 = np.zeros((c,1))*0.01
```

```python
#Activation Functions

def relu(z):
    return np.maximum(0,z)

def sigmoid(z):
    return 1/(1+np.exp(-z))

def softmax(z):
    exps = np.exp(z - np.max(z, axis=0, keepdims=True))
    return exps / np.sum(exps, axis=0, keepdims=True)

def tanh(z):
    return np.tanh(z)
```

```python
def forwardPropagation(W1,b1,W2,b2,X,Y):
    z1=np.dot(W1,X)+b1
    a1=tanh(z1)
    z2=np.dot(W2,a1)+b2
    a2=y_hat=softmax(z2)

    return (z1,a1,z2,a2)

```

```python

def Loss(y_hat,i,Y):

    Loss_function=-1/m*np.sum(Y.T*np.log(y_hat)) #Simple Multiplication so we want both the array to have same (m X n)-(m X n) dimensions

    print(f"Epoch {i+1}: Loss = {Loss_function}")

    return Loss_function
```

```python
#Backpropagation


def backpropagation(W2,b2,W1,b1,a1 ,X,Y, y_hat,lr):
    #1
    dL_dz2=1/m*(y_hat-Y.T)
    dL_dw2= np.dot(dL_dz2,a1.T)#Vector
    dL_db2= np.sum(dL_dz2,axis=1, keepdims=True)#we don't want scaler we want something 10x1 because each neuron will have its bias

    #hidden
    dL_da1=np.dot(W2.T,dL_dz2)
    dL_dz1=dL_da1*(1-np.power(a1,2))#Tanh wants element wise multiplication so no dot product

    #2
    dL_dw1=np.dot(dL_dz1,X.T)
    dL_db1=np.sum(dL_dz1,axis=1,keepdims=True)#same here we want bias for all

    #Updation



    W2=W2-(lr*dL_dw2)
    b2=b2-(lr*dL_db2)

    W1=W1-(lr*dL_dw1)
    b1=b1-(lr*dL_db1)

    return (W2,b2),(W1,b1)



```

```python
#Gradient Descent

epochs=10

learning_rate=15

for i in range(epochs):
    z1, a1, z2, y_hat = forwardPropagation(W1,b1,W2,b2,X,Y)

    Loss(y_hat,i,Y)

    (W2,b2),(W1,b1) = backpropagation(
        W2,b2,W1,b1,a1,X,Y,y_hat,learning_rate
    )
```

# Stochastic Gradient Descent

```python
def Loss_SGD(y_hat, Y_true, epoch):
    """
    For GD:
        Y_true shape = (10, m)

    For SGD:
        Y_true shape = (10, 1)
    """

    loss = -np.sum(Y_true * np.log(y_hat + 1e-8))

    print(f"Epoch {epoch + 1}: Loss = {loss}")

    return loss
```

```python
def backpropagation_SGD(W2, b2, W1, b1, a1, X, Y, y_hat, lr):
    """
    Y shape = (10, 1)
    Single sample only
    """

    # Output layer
    dL_dz2 = y_hat - Y   # NO Y.T and NO 1/m becoz it has only one value so 1/1=1

    dL_dW2 = np.dot(dL_dz2, a1.T)
    dL_db2 = np.sum(dL_dz2, axis=1, keepdims=True)

    # Hidden layer
    dL_da1 = np.dot(W2.T, dL_dz2)

    dL_dz1 = dL_da1 * (1 - np.power(a1, 2))

    # First layer
    dL_dW1 = np.dot(dL_dz1, X.T)
    dL_db1 = np.sum(dL_dz1, axis=1, keepdims=True)

    # Update
    W2 = W2 - lr * dL_dW2
    b2 = b2 - lr * dL_db2

    W1 = W1 - lr * dL_dW1
    b1 = b1 - lr * dL_db1

    return (W2, b2), (W1, b1)
```

```python
epochs = 10
learning_rate = 0.1

for i in range(epochs):

    for j in range(X.shape[1]):   # loop over samples

        x = X[:, j].reshape(-1, 1)   # (200,1)
        y = Y[j].reshape(-1, 1)      # (10,1)

        z1, a1, z2, y_hat = forwardPropagation(
            W1, b1, W2, b2, x,y
        )

        (W2, b2), (W1, b1) = backpropagation_SGD(
            W2, b2, W1, b1,
            a1, x, y, y_hat,
            learning_rate
        )

    # print loss once per epoch
    Loss_SGD(y_hat, y, i)
```

# Mini-Batch Gradient Descent


```python
def Loss_MGD(y_hat,y_true,epoch):
    """
    For GD:
        Y_true shape = (10, m)

    For SGD:
        Y_true shape = (10, 1)

    For MGD
        Y_true shape=(10,batch_size)
    """

    loss = -np.sum(y_true * np.log(y_hat + 1e-8))

    print(f"Epoch {epoch + 1}: Loss = {loss}")

    return loss
```

```python
def backpropagation_MGD(W2, b2, W1, b1, a1, X, Y, y_hat, lr,batch_size):
    """
    Y shape = (10, batch_size)
    batch_size sample only
    """

    # Output layer
    dL_dz2 =1/batch_size*( y_hat - Y)   # NO Y.T and NO 1/batch_size

    dL_dW2 = np.dot(dL_dz2, a1.T)
    dL_db2 = np.sum(dL_dz2, axis=1, keepdims=True)

    # Hidden layer
    dL_da1 = np.dot(W2.T, dL_dz2)

    dL_dz1 = dL_da1 * (1 - np.power(a1, 2))

    # First layer
    dL_dW1 = np.dot(dL_dz1, X.T)
    dL_db1 = np.sum(dL_dz1, axis=1, keepdims=True)

    # Update
    W2 = W2 - lr * dL_dW2
    b2 = b2 - lr * dL_db2

    W1 = W1 - lr * dL_dW1
    b1 = b1 - lr * dL_db1

    return (W2, b2), (W1, b1)
```

```python
epochs = 10
learning_rate = 0.1

for i in range(epochs):
    batch_size=(int)(X.shape[1]/10)

    for j in range(0,X.shape[1],batch_size):   # loop over samples

        x = X[:, j:j+batch_size]
        y = Y[j:j+batch_size].T

        z1, a1, z2, y_hat = forwardPropagation(
            W1, b1, W2, b2, x,y
        )

        (W2, b2), (W1, b1) = backpropagation_MGD(
            W2, b2, W1, b1,
            a1, x, y, y_hat,
            learning_rate,
            batch_size
        )

    # print loss once per epoch
    Loss_MGD(y_hat, y, i)
```
