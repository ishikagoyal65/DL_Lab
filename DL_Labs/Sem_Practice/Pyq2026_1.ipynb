```python
y_labels=np.random.randint(0,c,size=m)# m no. of outputs each having values in 0 to 5 classes
y=np.zeros((m,c))
y[np.arange(m),y_labels]=1 #m outputs me jitne bhi labels h unko 1 mark krdo y me jo ki ek zeros ka array h 
```

```python
W1=np.random.randn(Dx,Da1)
b1=np.zeros((Da1,1))

W2=np.random.randn(Da1,Da2)
b2=np.zeros((Da2,1))

W3=np.random.randn(Da2,c)
b3=np.zeros((c,1))
```

```python
#Activation Functions

def relu(z):
    return np.maximum(0,z)

def tanh(z):
    return np.tanh(z)

def sigmoid(z):
    return 1/(1+np.exp(-z))

def softmax(z):
    exps = np.exp(z - np.max(z, axis=0, keepdims=True))
    return exps / np.sum(exps, axis=0, keepdims=True)
```

```python
# Forward Propagation

def Forward_Propagation(W1,b1,W2,b2,W3,b3,X,y):
    z1=np.dot(W1.T,X)+b1
    a1=relu(z1)
    z2=np.dot(W2.T,a1)+b2
    a2=relu(z2)
    z3=np.dot(W3.T,a2)+b3
    a3=y_hat=softmax(z3)

    return (z1,a1,z2,a2,z3,a3)
```

```python
def Loss_function(y_hat,m,y,epoch):
    loss=(-1/m)*np.sum(y.T*np.log(y_hat+1e-9))
    print(f"For epoch {epoch} Loss ={loss}")

    return loss
```

```python
def BackwardPropagation(z1,a1,z2,a2,W1,b1,W2,b2,W3,b3,y_hat,lr):
    #1
    dL_dz3=(1/m)*(y_hat-y.T)

    dL_dw3=np.dot(a2,dL_dz3.T)
    dL_db3=np.sum(dL_dz3,axis=1,keepdims=True)

    #2
    dL_da2=np.dot(W3,dL_dz3)
    dL_dz2=dL_da2*(z2>0)

    dL_dw2=np.dot(a1,dL_dz2.T)
    dL_db2=np.sum(dL_dz2,axis=1,keepdims=True)
    
    #3
    dL_da1=np.dot(W2,dL_dz2)
    dL_dz1=dL_da1*(z1>0)

    dL_dw1=np.dot(X,dL_dz1.T)
    dL_db1=np.sum(dL_dz1,axis=1,keepdims=True)

    W3=W3-(lr*dL_dw3)
    b3=b3-(lr*dL_db3)

    W2=W2-(lr*dL_dw2)
    b2=b2-(lr*dL_db2)

    W1=W1-(lr*dL_dw1)
    b1=b1-(lr*dL_db1)

    return (W3,b3,W2,b2,W1,b1)


```

```python
epochs=10
learning_rate=0.5

for i in range(epochs):
    (z1,a1,z2,a2,z3,a3)=Forward_Propagation(W1,b1,W2,b2,W3,b3,X,y)
    Loss_function(a3,m,y,i)
 
    (W3,b3,W2,b2,W1,b1)=BackwardPropagation(z1,a1,z2,a2,W1,b1,W2,b2,W3,b3,a3,learning_rate)
 
```
